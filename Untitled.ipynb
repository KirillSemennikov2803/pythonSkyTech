```python
import numpy as np
from PIL import Image
```

```python
img = Image.open('scale16.png').convert('RGBA')
arr = np.array(img)
arrOut = arr
counti = 0
countj = 0
prefElem = arr[1][1]
```

```python
import os

def cls():
    os.system(['clear','cls'][os.name == 'nt'])
```

```python
for i in arr:
    countj=0
    for j in i:
        #print(countj)
        if(np.array_equal(j,[250,248,240,255]) or np.array_equal(j,[237,236,238,255])):
            arrOut[counti][countj]=[130,130,130,255]
        elif(np.array_equal(j,[223,240,206,255]) or np.array_equal(j,[185,214,229,255]) or np.array_equal(j,[219,229,221,255])
             or np.array_equal(j,[252,200,132,255])):
            arrOut[counti][countj]=[255,255,255,255]
        elif(np.array_equal(j,[244,239,227,255]) or np.array_equal(j,[227,227,229,255])):
            arrOut[counti][countj]=[71,71,71,255]
        else:
            #print(countj)
            arrOut[counti][countj]=prefElem
        prefElem = j
        countj +=1
        if (countj >=5248):
            countj = 5248
    counti+=1
    cls()
    print(counti)
    if (counti >=5248):
            counti = 5248
```

```python
img = Image.fromarray(arrOut, 'RGB')
img.save('my.png')
img.show()
```
